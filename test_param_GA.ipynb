```python
import numpy as np 
from test_functions import *
from genetic_algorithm import GeneticAlgorithm
from matplotlib import pyplot as plt
```

```python
bound=[-100,100]
d=15
bounds=[bound for i in range(d)]
model=GeneticAlgorithm(bound=bounds,iteration_num=5000,dna_size=15,
                      evaluator=sphere,population_size=100,selection_strategy='rank')
min_fitness=model.fitness['best'][-1]
print('best fitness=',min_fitness,'\n best_solution=',model.fitness['best_solution'][-1])
ax1=plt.subplot(111)
ax1.plot(np.arange(len(model.fitness['best'])),model.fitness['best'],'b',label='best')
ax1.plot(np.arange(len(model.fitness['best'])),model.fitness['mean'],'r',label='mean')
ax1.set_title('Sphere on [-100,100]')
ax1.set_xlabel('iteration times')
ax1.set_ylabel('fitness(function value)')
ax1.legend()
plt.show()
```
